# bắt đầu cào post từ 1 page của Facebook

```python
FANPAGE_LINK ="HoiMePhimAnh"
FOLDER_PATH = "/content/drive/MyDrive/Machine_learing"
COOKIE_PATH = "/content/drive/MyDrive/Machine_learning/cookies.txt"

PAGES_NUMBER = 5 # số trang cần cào
```

```python
from facebook_scraper import get_posts
import pandas as pd
```

```python
post_list = []
for post in get_posts(FANPAGE_LINK,
                    options={ "allow_extra_requests": True},
                    extra_info=True, pages=10, cookies=COOKIE_PATH):
    post_list.append(post)
```

```python
# Initialize dataframe to scrape Facebook post
post_df_full = pd.DataFrame(columns = [])
# Start to collect Facebook post data by facebook_scraper library
for post in post_list:
    post_entry = post
    fb_post_df = pd.DataFrame.from_dict(post_entry, orient='index')
    fb_post_df = fb_post_df.transpose()
    post_df_full = post_df_full.append(fb_post_df)
    print(post['post_id']+' get')
```
